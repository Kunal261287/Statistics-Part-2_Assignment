1. What is hypothesis testing in statistics?
   --> Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves testing an assumption (called a hypothesis) about a population parameter.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
   --> The null hypothesis is a statement that there is no effect, no difference, or no relationship in the population. It represents the status quo or a baseline assumption. It is the hypothesis that we test against using data.

3. What is the significance level in hypothesis testing, and why is it important?
   --> The significance level, denoted by Î± (alpha), is a threshold used to determine whether to reject the null hypothesis (Hâ‚€) in a hypothesis test. It represents the probability of making a Type I error â€” that is, rejecting a true null hypothesis.

4. What does a P-value represent in hypothesis testing?
   --> The P-value is the probability of obtaining a result as extreme as, or more extreme than, the one observed in your sample data, assuming that the null hypothesis is true.

5. How do you interpret the P-value in hypothesis testing?
   --> The P-value tells you how likely your observed results (or more extreme ones) are if the null hypothesis (Hâ‚€) were actually true.

   --> The Z-test is a statistical test used to determine whether there's a significant difference between sample and population means (or between two sample means), assuming the population variance is known and the data is approximately normally distributed.

```python
#2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# 1. Define the range of the discrete uniform distribution
low_bound = 1
high_bound = 6  # Represents a standard six-sided die

# Create an array of possible outcomes
outcomes = np.arange(low_bound, high_bound + 1)

# 2. Calculate the Probability Mass Function (PMF)
# For a discrete uniform distribution, each outcome has an equal probability.
# The probability is 1 / (number of outcomes).
pmf = randint.pmf(outcomes, low=low_bound, high=high_bound + 1)

# 3. Plot the PMF
plt.figure(figsize=(8, 5))
plt.stem(outcomes, pmf, use_line_collection=True, markerfmt='o', basefmt=' ')
plt.title('Probability Mass Function (PMF) of a Discrete Uniform Distribution')
plt.xlabel('Possible Outcomes')
plt.ylabel('Probability')
plt.xticks(outcomes) # Ensure x-axis ticks align with outcomes
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, max(pmf) + 0.05) # Adjust y-axis limit for better visualization
plt.show()
```

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    """
    Calculate the probability mass function of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1)
    p (float): Probability of success (0 <= p <= 1)

    Returns:
    float: The probability P(X = x)
    """
    if x == 0:
        return 1 - p
    elif x == 1:
        return p
    else:
        return 0  # For invalid values of x
```

```python
'''6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution'''
import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_cdf(a, b):
    """
    Calculate the CDF values of a discrete uniform distribution from a to b.
    
    Parameters:
    a (int): lower bound (inclusive)
    b (int): upper bound (inclusive)
    
    Returns:
    x (np.array): possible values
    cdf (np.array): corresponding CDF values
    """
    x = np.arange(a, b + 1)
    prob = 1 / (b - a + 1)
    pmf = np.full_like(x, prob, dtype=float)
    cdf = np.cumsum(pmf)
    return x, cdf

# Parameters
a = 1
b = 6

# Calculate CDF
x, cdf = discrete_uniform_cdf(a, b)

# Plotting
plt.step(x, cdf, where='post', label='CDF')
plt.scatter(x, cdf, color='red')  # show points for clarity
plt.title(f'Cumulative Distribution Function (CDF) of Discrete Uniform [{a}, {b}]')
plt.xlabel('Value')
plt.ylabel('CDF')
plt.xticks(x)
plt.ylim(0, 1.05)
plt.grid(True)
plt.legend()
plt.show()
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    """
    Calculate Z-scores from the input data and plot them.

    Parameters:
    data (list or np.array): Input dataset (numeric values)
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation
    
    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(8, 4))
    plt.plot(z_scores, marker='o', linestyle='-', color='purple')
    plt.axhline(0, color='gray', linestyle='--')
    plt.title('Z-scores of the Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.grid(True)
    plt.show()
    
    return z_scores

# Example usage
sample_data = [10, 12, 9, 15, 13, 8, 14, 11, 7, 12]
z_scores = calculate_and_plot_zscores(sample_data)
print("Z-scores:", z_scores)
```

```python
'''14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation
    z = (data_point - mean) / std
    return z

def plot_standard_normal_with_z(z):
    x = np.linspace(-4, 4, 1000)
    pdf = norm.pdf(x, 0, 1)

    plt.plot(x, pdf, label='Standard Normal PDF')
    plt.fill_between(x, 0, pdf, where=(x <= z), color='lightblue', alpha=0.5,
                     label=f'Area â‰¤ Z = {z:.2f}')
    plt.axvline(z, color='red', linestyle='--', label=f'Z-score = {z:.2f}')
    plt.title('Standard Normal Distribution with Z-score')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example data and point
data = [10, 12, 9, 15, 13, 8, 14, 11, 7, 12]
data_point = 15

z_score = calculate_z_score(data_point, data)
print(f'Z-score of data point {data_point}: {z_score:.2f}')

plot_standard_normal_with_z(z_score)
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform a two-tailed Z-test on the sample.
    
    Parameters:
        sample (list or np.array): Sample data
        pop_mean (float): Hypothesized population mean (H0)
        pop_std (float): Known population standard deviation
        alpha (float): Significance level (default 0.05)
        
    Returns:
        z_stat (float): Calculated Z-statistic
        p_value (float): Two-tailed p-value
        reject_null (bool): Whether to reject the null hypothesis
    """
    n = len(sample)
    sample_mean = np.mean(sample)
    
    # Calculate standard error
    std_error = pop_std / np.sqrt(n)
    
    # Calculate Z-statistic
    z_stat = (sample_mean - pop_mean) / std_error
    
    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    
    # Decision: reject null if p-value < alpha
    reject_null = p_value < alpha
    
    return z_stat, p_value, reject_null

# Example data
sample_data = [52, 55, 47, 50, 49, 51, 48, 53, 54, 46]

# Known population parameters
population_mean = 50
population_std = 5  # assume known

# Significance level
alpha = 0.05

z_stat, p_val, reject = z_test(sample_data, population_mean, population_std, alpha)

print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_val:.4f}")
print("Reject null hypothesis:" if reject else "Fail to reject null hypothesis.")
```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of the data.
    
    Parameters:
    data (list or np.array): Sample data
    confidence (float): Confidence level (default 0.95)
    
    Returns:
    (float, float): Lower and upper bounds of the confidence interval
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # standard error of the mean
    
    # t critical value for two-tailed test
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)
    
    margin_of_error = t_crit * std_err
    
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    
    return lower_bound, upper_bound

# Example data
sample_data = [52, 55, 47, 50, 49, 51, 48, 53, 54, 46]

# Calculate 95% confidence interval
conf_level = 0.95
lower, upper = confidence_interval(sample_data, conf_level)

print(f"{int(conf_level*100)}% Confidence interval: ({lower:.2f}, {upper:.2f})")
print(f"Interpretation: We are {int(conf_level*100)}% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")
```

```python
#Practical Part - 2
'''Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.'''
import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform a two-tailed Z-test.
    
    Parameters:
        sample (array-like): sample data
        pop_mean (float): population mean under null hypothesis
        pop_std (float): known population standard deviation
        alpha (float): significance level (default 0.05)
        
    Returns:
        z_stat (float): computed Z statistic
        p_value (float): two-tailed p-value
        reject_null (bool): whether to reject the null hypothesis
    """
    n = len(sample)
    sample_mean = np.mean(sample)
    std_error = pop_std / np.sqrt(n)
    
    z_stat = (sample_mean - pop_mean) / std_error
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))  # two-tailed
    
    reject_null = p_value < alpha
    
    print(f"Sample Mean = {sample_mean:.3f}")
    print(f"Z-statistic = {z_stat:.3f}")
    print(f"P-value = {p_value:.4f}")
    
    if reject_null:
        print(f"Reject the null hypothesis at alpha = {alpha}")
        print("Conclusion: Sample mean is significantly different from the population mean.")
    else:
        print(f"Fail to reject the null hypothesis at alpha = {alpha}")
        print("Conclusion: No significant difference between sample mean and population mean.")
    
    return z_stat, p_value, reject_null

# Example usage:
sample_data = [52, 55, 47, 50, 49, 51, 48, 53, 54, 46]
population_mean = 50
population_std = 5  # known population standard deviation
alpha_level = 0.05

z_test(sample_data, population_mean, population_std, alpha_level)
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test_with_plot(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    std_error = pop_std / np.sqrt(n)

    # Calculate Z statistic
    z_stat = (sample_mean - pop_mean) / std_error

    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Population Mean: {pop_mean}")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Determine rejection regions
    z_critical = norm.ppf(1 - alpha/2)
    reject_left = -z_critical
    reject_right = z_critical

    if abs(z_stat) > z_critical:
        decision = "Reject the null hypothesis."
    else:
        decision = "Fail to reject the null hypothesis."

    print(decision)

    # Plot the standard normal distribution
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10,6))
    plt.plot(x, y, label='Standard Normal Distribution')

    # Shade rejection regions
    x_reject_left = np.linspace(-4, reject_left, 200)
    x_reject_right = np.linspace(reject_right, 4, 200)
    plt.fill_between(x_reject_left, 0, norm.pdf(x_reject_left), color='red', alpha=0.3, label='Rejection Region')
    plt.fill_between(x_reject_right, 0, norm.pdf(x_reject_right), color='red', alpha=0.3)

    # Mark critical values
    plt.axvline(reject_left, color='red', linestyle='--', label=f'Critical Values Â±{z_critical:.2f}')
    plt.axvline(reject_right, color='red', linestyle='--')

    # Mark observed Z statistic
    plt.axvline(z_stat, color='blue', linestyle='-', linewidth=2, label=f'Observed Z = {z_stat:.2f}')

    plt.title('Two-tailed Z-test Decision Regions')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
np.random.seed(123)
sample = np.random.normal(loc=52, scale=5, size=40)
population_mean = 50
population_std = 6
alpha_level = 0.05

two_tailed_z_test_with_plot(sample, population_mean, population_std, alpha_level)
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(mu0=0, mu1=1, sigma=1, n=30, alpha=0.05):
    """
    Visualize Type 1 and Type 2 errors for a one-sided test:
    H0: mean = mu0 vs H1: mean = mu1 (mu1 > mu0)
    
    Parameters:
        mu0 (float): mean under null hypothesis
        mu1 (float): mean under alternative hypothesis
        sigma (float): population standard deviation
        n (int): sample size
        alpha (float): significance level (Type 1 error rate)
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for one-tailed test (right tail)
    z_critical = norm.ppf(1 - alpha)
    crit_value = mu0 + z_critical * se

    # x range for plotting
    x_min = mu0 - 4*se
    x_max = mu1 + 4*se
    x = np.linspace(x_min, x_max, 1000)

    # Distributions under H0 and H1
    pdf_h0 = norm.pdf(x, loc=mu0, scale=se)
    pdf_h1 = norm.pdf(x, loc=mu1, scale=se)

    plt.figure(figsize=(10,6))

    # Plot H0 distribution
    plt.plot(x, pdf_h0, label='Null Hypothesis (H0)', color='blue')
    # Plot H1 distribution
    plt.plot(x, pdf_h1, label='Alternative Hypothesis (H1)', color='green')

    # Shade Type 1 error (Î±): area in H0 distribution beyond critical value
    x_type1 = np.linspace(crit_value, x_max, 500)
    plt.fill_between(x_type1, 0, norm.pdf(x_type1, mu0, se), color='red', alpha=0.3, label='Type 1 Error (Î±)')

    # Shade Type 2 error (Î²): area in H1 distribution left of critical value
    x_type2 = np.linspace(x_min, crit_value, 500)
    plt.fill_between(x_type2, 0, norm.pdf(x_type2, mu1, se), color='orange', alpha=0.3, label='Type 2 Error (Î²)')

    # Plot critical value line
    plt.axvline(crit_value, color='black', linestyle='--', label=f'Critical Value = {crit_value:.2f}')

    plt.title('Visualization of Type 1 and Type 2 Errors')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
plot_type1_type2_errors(mu0=50, mu1=52, sigma=10, n=40, alpha=0.05)
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a sample mean using the t-distribution.

    Parameters:
        data (array-like): Sample data
        confidence (float): Confidence level (default is 0.95)

    Returns:
        float: Margin of error
    """
    n = len(data)
    sample_std = np.std(data, ddof=1)
    std_error = sample_std / np.sqrt(n)
    
    # t critical value
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)
    
    margin_of_error = t_critical * std_error
    
    print(f"Sample Size: {n}")
    print(f"Sample Standard Deviation: {sample_std:.2f}")
    print(f"Standard Error: {std_error:.2f}")
    print(f"T Critical Value (df={n-1}): {t_critical:.4f}")
    print(f"Margin of Error at {int(confidence*100)}% confidence: Â±{margin_of_error:.2f}")
    
    return margin_of_error

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=50, scale=10, size=25)
calculate_margin_of_error(sample_data, confidence=0.95)
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayes_theorem(prior_H, prob_E_given_H, prob_E_given_not_H):
    """
    Apply Bayes' Theorem to calculate P(H|E)

    Parameters:
        prior_H (float): Prior probability of hypothesis H (P(H))
        prob_E_given_H (float): Likelihood P(E|H)
        prob_E_given_not_H (float): Likelihood P(E|Â¬H)

    Returns:
        float: Posterior probability P(H|E)
    """
    # Complement of H
    prior_not_H = 1 - prior_H

    # Total probability of evidence P(E)
    prob_E = (prob_E_given_H * prior_H) + (prob_E_given_not_H * prior_not_H)

    # Posterior probability P(H|E)
    posterior = (prob_E_given_H * prior_H) / prob_E

    print(f"Prior: P(H) = {prior_H}")
    print(f"Likelihood: P(E|H) = {prob_E_given_H}")
    print(f"P(E|Â¬H) = {prob_E_given_not_H}")
    print(f"Total Evidence P(E) = {prob_E:.4f}")
    print(f"Posterior: P(H|E) = {posterior:.4f}")
    
    return posterior

# Example values from disease testing
prior_disease = 0.01                # P(D)
prob_pos_given_disease = 0.99       # P(Pos | D)
prob_pos_given_no_disease = 0.05    # P(Pos | Â¬D)

bayes_theorem(prior_disease, prob_pos_given_disease, prob_pos_given_no_disease)
```

```python
#12. Perform a Chi-square test for independence between two categorical variables in Python.
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create a contingency table
# Rows: Gender (Male, Female)
# Columns: Product Preference (Like, Neutral, Dislike)

data = np.array([
    [30, 10, 5],   # Male
    [25, 20, 15]   # Female
])

# Optional: Convert to a DataFrame for readability
table = pd.DataFrame(data, 
                     columns=["Like", "Neutral", "Dislike"], 
                     index=["Male", "Female"])

print("Contingency Table:")
print(table)

# Step 2: Perform Chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(data)

# Step 3: Output results
print("\nChi-square Statistic:", round(chi2_stat, 4))
print("Degrees of Freedom:", dof)
print("P-value:", round(p_value, 4))
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, columns=table.columns, index=table.index))

# Step 4: Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nConclusion: Reject the null hypothesis.")
    print("â†’ There is a significant association between gender and product preference.")
else:
    print("\nConclusion: Fail to reject the null hypothesis.")
    print("â†’ No significant association between gender and product preference.")
```

```python
'''13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data.'''
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test of independence.

    Parameters:
        observed (2D array-like): The observed frequency table (contingency table)

    Returns:
        pd.DataFrame: Expected frequencies with the same shape as input
    """
    observed = np.array(observed)
    row_totals = observed.sum(axis=1).reshape(-1, 1)
    col_totals = observed.sum(axis=0).reshape(1, -1)
    total = observed.sum()

    # Expected frequency formula: (row total * column total) / grand total
    expected = (row_totals @ col_totals) / total
    return pd.DataFrame(expected, columns=["Col1", "Col2", "Col3"], index=["Row1", "Row2"])

# Example observed data (2x3 contingency table)
observed_data = [
    [20, 30, 25],  # Row 1
    [15, 35, 20]   # Row 2
]

# Calculate expected frequencies
expected_df = calculate_expected_frequencies(observed_data)

print("Observed Frequencies:")
print(pd.DataFrame(observed_data, columns=["Col1", "Col2", "Col3"], index=["Row1", "Row2"]))
print("\nExpected Frequencies:")
print(expected_df.round(2))
```

```python
#14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
import numpy as np
from scipy.stats import chisquare

# Step 1: Observed data (from rolling a die 60 times)
observed = np.array([8, 9, 10, 11, 12, 10])  # frequencies of faces 1 through 6

# Step 2: Expected frequencies (for a fair die)
total_rolls = observed.sum()
expected = np.array([total_rolls / 6] * 6)  # uniform distribution

# Step 3: Perform Chi-square goodness-of-fit test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Step 4: Print results
print("Observed:", observed)
print("Expected:", expected.astype(int))
print(f"\nChi-square Statistic: {chi_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 5: Interpret result
alpha = 0.05
if p_value < alpha:
    print("â†’ Reject the null hypothesis: The observed distribution differs from the expected.")
else:
    print("â†’ Fail to reject the null hypothesis: No significant difference from the expected distribution.")
```
